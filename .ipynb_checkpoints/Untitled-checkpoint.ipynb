```python
# import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
%matplotlib inline
```

## Read Data

```python
test_conditions=pd.read_csv('Python Test_conditions')
test_attempts=pd.read_csv('Python Test Files - Sudoku_attemps')
```

```python
test_conditions.shape
```

```python
test_attempts.shape
```

```python
game_attempts=test_attempts.groupby('Game ID').sum()
game_attempts['Completion Ratio']=game_attempts['Completed']/game_attempts['Tried']
game_attempts
```

```python
condition_attempts=test_conditions.set_index('Batch ID').join(test_attempts.groupby('Batch ID').sum(),how='inner')
condition_attempts['Completion Ratio']=condition_attempts['Completed']/condition_attempts['Tried']
condition_attempts
```

```python
game_condition_attempts=test_conditions.set_index('Batch ID').join(test_attempts.set_index('Batch ID'),how='inner')
game_condition_attempts['Completion Ratio']=game_condition_attempts['Completed']/game_condition_attempts['Tried']
game_condition_attempts
```

```python
plt.rcParams.update({'font.size': 20})
plt.pie([test_attempts['Completed'].sum(),test_attempts['Tried'].sum()-test_attempts['Completed'].sum()], labels=['Completed','Not completed'],
autopct='%1.1f%%', shadow=True)
plt.show()
```

* 52.7% attempts were successfully completed

```python
plt.rcParams['figure.figsize'] = [17, 5]
plt.rcParams.update({'font.size': 6})
success_variation_per_game=plt.bar(game_attempts.index, game_attempts['Completion Ratio']*100,edgecolor='black')
plt.ylabel('Percentage')
plt.xlabel('Games')
plt.title('Completion Percentage for each game')
plt.show()
```

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
y_val= game_condition_attempts['Completed']
x_data=game_condition_attempts.drop("Completed",axis=1)
X_train, X_eval,y_train,y_eval=train_test_split(x_data,y_val,test_size=0.3,random_state=101)
```

```python
feature_cols=[]
for cols in game_condition_attempts.columns[:-2]:
    column=tf.feature_column.numeric_column(cols)
    feature_cols.append(column)
print(feature_cols)
```

```python
#The estimator model
model=tf.estimator.DNNRegressor(hidden_units=[6,10,6],feature_columns=feature_cols)
#the input function
input_func=tf.estimator.inputs.pandas_input_fn(X_train,y_train,batch_size=10,num_epochs=1000,shuffle=True)
```
